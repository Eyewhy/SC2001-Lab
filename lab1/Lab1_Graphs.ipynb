```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV files
test_a = pd.read_csv("test_a.csv")
test_s = pd.read_csv("test_s.csv")
test_x = pd.read_csv("test_x.csv")

# Separate columns for hybrid and merge sort times and comparisons
hybrid_time_columns = [col for col in test_a.columns if "hybrid" in col]
merge_time_columns = [col for col in test_a.columns if "merge" in col]
hybrid_comparison_columns = [col for col in test_a.columns if "hcomps" in col]
merge_comparison_columns = [col for col in test_a.columns if "mcomps" in col]

# Compute average for hybrid and merge sort times
hybrid_time_avg = test_a[hybrid_time_columns].mean(axis=1)
merge_time_avg = test_a[merge_time_columns].mean(axis=1)
plt.figure(figsize=(10, 6))
plt.plot(test_a[test_a.columns[0]], hybrid_time_avg, marker='o', label='Hybrid Sort Avg Time')
plt.plot(test_a[test_a.columns[0]], merge_time_avg, marker='o', label='Merge Sort Avg Time')
plt.xlabel(test_a.columns[0])
plt.ylabel("Average Time")
plt.title("Average Sorting Time vs Input Size (test_a.csv)")
plt.xscale("log")
plt.yscale("log")
plt.legend()
plt.grid()
plt.show()

# Compute average for hybrid and merge sort comparisons
hybrid_comparison_avg = test_a[hybrid_comparison_columns].mean(axis=1)
merge_comparison_avg = test_a[merge_comparison_columns].mean(axis=1)
plt.figure(figsize=(10, 6))
plt.plot(test_a[test_a.columns[0]], hybrid_comparison_avg, marker='o', label='Hybrid Sort Avg Comparisons')
plt.plot(test_a[test_a.columns[0]], merge_comparison_avg, marker='o', label='Merge Sort Avg Comparisons')
plt.xlabel(test_a.columns[0])
plt.ylabel("Average Comparisons")
plt.title("Average Comparisons vs Input Size (test_a.csv)")
plt.xscale("log")
plt.yscale("log")
plt.legend()
plt.grid()
plt.show()
```

```python
# Compute average runtime for test_s.csv
runtime_columns = [col for col in test_s.columns if "run" in col]
avg_runtime = test_s[runtime_columns].mean(axis=1)
plt.figure(figsize=(10, 6))
plt.plot(test_s[test_s.columns[0]], avg_runtime, marker='o', label='Average Runtime')
plt.xlabel("s")
plt.ylabel("Average Run Time")
plt.title("Average Runtime vs s (test_s.csv)")
plt.legend()
plt.grid()
plt.show()

# Compute average comparisons for test_s.csv
comparison_columns = [col for col in test_s.columns if "comps" in col]
avg_comparisons = test_s[comparison_columns].mean(axis=1)
plt.figure(figsize=(10, 6))
plt.plot(test_s[test_s.columns[0]], avg_comparisons, marker='o', label='Average Comparisons')
plt.xlabel("s")
plt.ylabel("Average Comparisons")
plt.title("Average Comparisons vs s (test_s.csv)")
plt.legend()
plt.grid()
plt.show()
```

```python
# Plot individual graphs for different a values in test_x.csv
unique_a_values = test_x[test_x.columns[1]].unique()
time_columns = [col for col in test_x.columns if "run" in col]
# del test_x['Unnamed: 12']
for a_val in unique_a_values:
    subset = test_x[test_x[test_x.columns[1]] == a_val]
    plt.figure(figsize=(10, 6))
    for i in range(2, len(test_x.columns), 2):
        plt.plot(subset[test_x.columns[0]], subset[test_x.columns[i]], label=test_x.columns[i])

    time_avg = subset[time_columns].mean(axis=1)
    plt.plot(subset[test_x.columns[0]], time_avg, label='Average Runtime')
    
    plt.xlabel("s")
    plt.ylabel("Run Time")
    plt.title(f"Performance of Runs vs s for a = {a_val} (test_x.csv)")
    plt.legend()
    plt.grid()
    plt.show()
```

```python

# Plot individual graphs for different a values in test_x.csv
test_x = pd.read_csv("test_x2.csv")

unique_a_values = test_x[test_x.columns[1]].unique()
time_columns = [col for col in test_x.columns if "run" in col]
cols = []
# del test_x['Unnamed: 12']
plt.figure(figsize=(10, 6))
for a_val in unique_a_values:
    subset = test_x[test_x[test_x.columns[1]] == a_val]
    


    time_avg = subset[time_columns].mean(axis=1)
    time_avg = time_avg/a_val
    cols.append(time_avg.reset_index(drop=True))
    plt.plot(subset[test_x.columns[0]], time_avg, label=a_val)
    

plt.xlabel("s")
plt.ylabel("Run Time/item")
plt.title(f"Performance of Runs vs s (test_x.csv)")
plt.legend()
plt.grid()
plt.show()
```

```python
avg = pd.DataFrame(cols)
plt.figure(figsize=(10, 6))
plt.plot(subset[test_x.columns[0]], avg.mean(), label="Average")
plt.xlabel("s")
plt.ylabel("Run Time/item")
plt.title(f"Performance of Runs vs s (test_x.csv)")
plt.legend()
plt.grid()

plt.show()
```
