```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
test = pd.read_csv('test.csv')
test_e = pd.read_csv('test_e.csv')
test_v = pd.read_csv('test_v.csv')

# Compute averages
test['array_avg'] = test[[f'a{i}' for i in range(1, 6)]].mean(axis=1)
test['pq_avg'] = test[[f'q{i}' for i in range(1, 6)]].mean(axis=1)

test_e['array_avg'] = test_e[[f'a{i}' for i in range(1, 6)]].mean(axis=1)
test_e['pq_avg'] = test_e[[f'q{i}' for i in range(1, 6)]].mean(axis=1)

test_v['array_avg'] = test_v[[f'a{i}' for i in range(1, 6)]].mean(axis=1)
test_v['pq_avg'] = test_v[[f'q{i}' for i in range(1, 6)]].mean(axis=1)

# Visualise performance for varying graph sizes
plt.figure(figsize=(10, 6))
plt.plot(test['i'], test['array_avg'], label='Array (avg)')
plt.plot(test['i'], test['pq_avg'], label='Priority Queue (avg)')
plt.title('Dijkstra Performance vs Graph Size')
plt.xlabel('Graph Size (Number of Vertices)')
plt.ylabel('Execution Time (ms)')
plt.legend()
plt.grid()
plt.show()

# Visualise performance for varying number of edges
plt.figure(figsize=(10, 6))
plt.plot(test_e['e'], test_e['array_avg'], label='Array (avg)')
plt.plot(test_e['e'], test_e['pq_avg'], label='Priority Queue (avg)')
plt.title('Dijkstra Performance vs Number of Edges')
plt.xlabel('Number of Edges')
plt.ylabel('Execution Time (ms)')
plt.legend()
plt.grid()
plt.show()

# Visualise performance for varying number of vertices
plt.figure(figsize=(10, 6))
plt.plot(test_v['v'], test_v['array_avg'], label='Array (avg)')
plt.plot(test_v['v'], test_v['pq_avg'], label='Priority Queue (avg)')
plt.title('Dijkstra Performance vs Number of Vertices')
plt.xlabel('Number of Vertices')
plt.ylabel('Execution Time (ms)')
plt.legend()
plt.grid()
plt.show()
```

```python

test_high_e = pd.read_csv('test_high_e.csv')
test_low_e = pd.read_csv('test_low_e.csv')

# Compute averages
test_high_e['array_avg'] = test_high_e[[f'a{i}' for i in range(1, 6)]].mean(axis=1)
test_high_e['pq_avg'] = test_high_e[[f'q{i}' for i in range(1, 6)]].mean(axis=1)

test_low_e['array_avg'] = test_low_e[[f'a{i}' for i in range(1, 6)]].mean(axis=1)
test_low_e['pq_avg'] = test_low_e[[f'q{i}' for i in range(1, 6)]].mean(axis=1)

labels = {
    'array_avg': 'Array',
    'pq_avg': 'Priority Queue',
}
for djtype in ['array_avg','pq_avg']:
    plt.figure(figsize=(10, 6))
    plt.plot(test_high_e['i'], test_high_e[djtype], label='High Edges (avg)')
    plt.plot(test_low_e['i'], test_low_e[djtype], label='Low Edges (avg)')
    plt.plot(test['i'], test[djtype], label='Medium Edges (avg)')
    plt.title(f'Dijkstra ({labels[djtype]}) Performance vs Edge Number')
    plt.xlabel('Number of Edges')
    plt.ylabel('Execution Time (ms)')
    plt.legend()
    plt.grid()
    plt.show()

labels2 = ['Medium', 'High', 'Low']
for i, etype in enumerate([test,test_high_e,test_low_e]):
    plt.figure(figsize=(10, 6))
    plt.plot(getattr(etype,'i'), getattr(etype,'array_avg'), label='Array (avg)')
    plt.plot(getattr(etype,'i'), getattr(etype,'pq_avg'), label='Priority Queue (avg)')
    plt.title(f'Dijkstra Performance vs Edge Number ({labels2[i]} scaling)')
    plt.xlabel('Number of Edges')
    plt.ylabel('Execution Time (ms)')
    plt.legend()
    plt.grid()
    plt.show()

plt.figure(figsize=(10, 6))
plt.plot(test_low_e['i'], test_low_e['pq_avg'], label='Low Edges (avg)')
plt.title(f'Dijkstra ({labels["pq_avg"]}) Performance vs Edge Number')
plt.xlabel('Number of Edges')
plt.ylabel('Execution Time (ms)')
plt.legend()
plt.grid()
plt.show()
```
